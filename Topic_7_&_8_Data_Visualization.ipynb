```python
# import package for visualization
import seaborn as sns
```

```python
# barplot (horizontal)
sns.barplot(data=train_df,x="Survived",y="Sex")
```

```python
# barplot (vertical)
sns.barplot(data=train_df,x="Sex",y="Survived")
```

```python
# distplot
import matplotlib.pyplot as plt 
data = pd.DataFrame(train_df['Survived'])
res = sns.distplot(data)
plt.show()
```

```python
# boxplot
sns.boxplot(data=train_df ['Fare'])
sns.set(rc={'figure.figsize':(10,10)})
```

```python
# scatterplot
sns.scatterplot(data=train_df,x="PassengerId",y="Fare")
```

```python
# pie chart
train_df

grouped_data=train_df.groupby('Sex').count()['PassengerId'].reset_index()
grouped_data=grouped_data.rename(columns={"Sex":"Gender","PassengerId": "Total"})

fig1, ax1 = plt.subplots()
ax1.pie(grouped_data['Total'], labels=grouped_data['Gender'], autopct='%1.1f%%', startangle=180)

plt.show()
```
